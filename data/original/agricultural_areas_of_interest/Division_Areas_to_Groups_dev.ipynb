```python
import pandas as pd
import random as rand
import numpy as np
```

```python
files_to_divide = ["Attendorn_dev.csv","Drolshagen_dev.csv","Finnentrop_dev.csv"]

#students_moodle = pd.read_csv(file_students_moodle)
#students_registered = pd.read_csv(file_students_registered)
```

```python
counter = 0
number_of_groups = 52
df_output = pd.DataFrame(columns = ["GroupID", "Attendorn_Ackerland","Attendorn_Grünland",
                                    "Drolshagen_Grünland","Drolshagen_Grünland",
                                   "Finnentrop_Grünland", "Finnentrop_Grünland"], index = range(number_of_groups))
dfLists = []
for file_to_divide in files_to_divide:
    #col_name = file_to_divide[-9] + "_"
    #print(col_name)
    try:
        df_complete = pd.read_csv(file_to_divide)
    except: 
        print("Could not find the table: ", file_to_divide)
    df_complete = df_complete[["Region_ID","bez"]]
    var_compare = "Ackerland"
    df_acker = df_complete.loc[df_complete["bez"]==var_compare]
    dfLists.append(df_acker)
    var_compare = "Grünland"
    df_gruen = df_complete.loc[df_complete["bez"]==var_compare]
    dfLists.append(df_gruen)
    
df_output = pd.DataFrame(columns = ["GroupID", "Attendorn_Ackerland","Attendorn_Grünland",
                                    "Drolshagen_Grünland","Drolshagen_Grünland",
                                   "Finnentrop_Grünland", "Finnentrop_Grünland"], index = range(number_of_groups))

counter = [0,0,0,0,0,0]
for n in range(number_of_groups):
    series_group = df_output.loc[n,:]
    try:
        series_group[0] = "Group_" + str(n+1)
    except:
        print("error while setting groupID")
    for it in range(len(dfLists)):
        try:
            series_group[it+1] = dfLists[it].iloc[counter[it],0]
            counter[it]+=1
        except:
            counter[it] = 0
            series_group[it+1] = dfLists[it].iloc[counter[it],0]
            counter[it]+=1
    
df_output


    
```

```python
df_output.to_csv("Regions_per_group.csv",header=False, index=False)
```
